```python

# Understanding the behavior of pitch control in blades with and without ice accretion is crucial for several reasons:
# 
# 1. Aerodynamic Performance: Ice accretion changes the shape of the blade, leading to a decrease in lift and an increase in drag. 
#    This affects the aerodynamic efficiency of the blade, which can significantly impact the overall performance of the turbine or aircraft.
# 
# 2. Power Production: For wind turbines, ice on the blades can lead to substantial power losses. Studies have shown that ice accretion 
#    can reduce power production by more than 50% during winter months. Monitoring and adjusting pitch control can help mitigate these 
#    losses by optimizing the blade angle to maintain as much efficiency as possible.
# 
# 3. Structural Integrity: Ice adds extra weight and can cause imbalances in the rotor. This can lead to increased vibrations and stress 
#    on the blade and other components, potentially causing damage or reducing the lifespan of the equipment. Proper pitch control can 
#    help manage these stresses by adjusting the blade angle to reduce the load.
# 
# 4. Safety: Ice shedding from blades can be hazardous. Proper pitch control can help minimize the risk by reducing the likelihood of 
#    ice buildup or by controlling the conditions under which ice is shed.
# 
# 5. Control System Optimization: Understanding how ice affects pitch control allows for the development of more effective control systems. 
#    These systems can automatically adjust the blade pitch to compensate for the presence of ice, maintaining optimal performance and safety.
# 
# By comparing the behavior of pitch control in blades with and without ice accretion, engineers can develop strategies to mitigate the 
# negative effects of ice, ensuring better performance, safety, and longevity of the blades.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants (improved values)
air_density = 1.341  # kg/m^3
wind_speed = 20.04  # m/s
air_temp = -10  # degrees Celsius (below freezing)
time_step = 1  # time step in seconds
total_time = 3600  # 1 hour simulation
liquid_water_content = 0.65 / 1000  # g/m^3 converted to kg/m^3
freezing_point = 0  # freezing point in Celsius
collection_efficiency = 0.315  # % of impinging water freezes
latent_heat_freezing = 334000  # J/kg for water

# Airfoil properties
chord_length = 1.0  # meter
span_length = 0.1  # meter
exposed_area = chord_length * span_length  # m^2

# Function to calculate heat transfer rate (convective heat loss)
def heat_transfer(air_temp, wind_speed, surface_temp):
    h_conv = 10.45 - wind_speed + 10 * np.sqrt(wind_speed)  # simplified convective heat transfer coefficient
    return h_conv * (air_temp - surface_temp)

# Function to calculate ice accretion
def ice_accretion(air_temp, wind_speed, liquid_water_content, time_elapsed):
    ice_thickness = 0  # initial ice thickness in meters
    surface_temp = 10  # assume surface starts at freezing point

    # Loop over time to simulate ice accretion
    for t in range(0, time_elapsed, time_step):
        if air_temp < freezing_point:
            # Water mass that can potentially freeze (kg/s)
            water_mass = liquid_water_content * air_density * wind_speed * exposed_area
            water_mass_freezing = water_mass * collection_efficiency  # Only a portion freezes
            
            # Ice mass accreted (kg)
            ice_mass = water_mass_freezing * time_step
            ice_volume = ice_mass / 1000  # kg to cubic meters
            
            # Calculate the thickness of ice formed (in meters)
            ice_thickness += ice_volume / exposed_area

    return ice_thickness

# Plot the ice accretion over time
time_steps = np.arange(0, total_time, time_step)
ice_growth = [ice_accretion(air_temp, wind_speed, liquid_water_content, t) for t in time_steps]

# Convert time to minutes for plotting
time_steps_minutes = time_steps / 60

plt.figure(figsize=(8, 8))
plt.plot(time_steps_minutes, ice_growth,label='Ice Thickness')
plt.title('Simplified Messinger Ice Accretion Model', fontsize=16, color=dtu_red)
plt.xlabel('Time [minutes]', fontsize=16)
plt.ylabel('Ice Thickness [m]', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(True, which='both', linestyle='--', linewidth=0.6, color='lightgray')
plt.legend(fontsize=16)
plt.tight_layout()
plt.show()

print(ice_growth[-1], ' meters')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
R = 89.15  # Rotor radius (m)
B = 3   # Number of blades
phi = np.radians(30)  # Azimuthal angle (degrees to radians)
r = np.linspace(0.1 * R, R, 100)  # Radial positions (avoid r=0)

# Define the Glauert Tip-Loss Correction factor
def glauert_tip_loss(B, R, r, phi):
    f = (B / 2) * (R - r) / (r * np.sin(phi))
    F = (2 / np.pi) * np.arccos(np.exp(-f))
    return F

# Axial induction factor (simplified) with and without correction
a_no_correction = 1 / (1 + 4 * np.pi * r / (B * R))  # Simplified model without correction
F = glauert_tip_loss(B, R, r, phi)  # Correction factor
a_with_correction = F * a_no_correction  # Adjusted with tip loss correction



# DTU Colors
dtu_blue = '#003C71'  # Example DTU blue color
dtu_red = '#990000'  # DTU red
dtu_grey = '#4C4C4C'  # DTU grey
dtu_white = '#FFFFFF'  # DTU white

# Plot the results
plt.figure(figsize=(8, 8))
plt.plot(r / R, a_no_correction, label='Without Tip-Loss Correction', linestyle='--')
plt.plot(r / R, a_with_correction, label='With Glauert Tip-Loss Correction', linestyle='-')
plt.xlabel('Normalized Radial Position ($r/R$)', fontsize=16)
plt.ylabel('Axial Induction Factor ($a$)', fontsize=16)
plt.title('Axial Induction Factor with and without Tip-Loss Correction', fontsize=16, color=dtu_red)
plt.legend(fontsize=16)
plt.grid(True, which='both', linestyle='--', linewidth=0.6, color='lightgray')
plt.tick_params(axis='both', which='major', labelsize=16)
plt.tight_layout()
plt.show()
```
